```python
import pandas as pd
mayone = pd.read_csv('20m.csv')
```

```python
mayone
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(9.5, 9.5))
axes.set_xticklabels(['Band0','Band 1','Band 2','Band 3','Band 4','Band 5','Band 6', 'Band 7'])
fig.suptitle('Figure PIN 20MAY18 Landsat8: Comparison of Satellite/GA-Field/CSIRO-Field data', fontweight='bold')
axes.set_ylabel('Reflectance')
plt.tight_layout(pad=3.5, w_pad=1.0, h_pad=1.0)
mayone.plot(y=['Sat_mean', 'Field_mean', 'CSIRO_Field_mean'], ax=axes, color=['black', 'blue', 'orange'])
plt.errorbar(x=mayone.index, y=mayone['Sat_mean'], yerr=mayone['Sat_SD'], color='black', capsize=3)
plt.errorbar(x=mayone.index, y=mayone['Field_mean'], yerr=mayone['Field_SD'], color='blue', capsize=3)
plt.errorbar(x=mayone.index, y=mayone['CSIRO_Field_mean'], yerr=mayone['CSIRO_Field_SD'], color='orange', capsize=3)
plt.savefig('FIG_20MAYcomp.png')
```
